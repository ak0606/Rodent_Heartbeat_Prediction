# **Data Visualization**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = '/kaggle/input/ecg-data/045_0-500000_corrected.csv'
data = pd.read_csv(path)
```

```python
# plt.figure(figsize = (12,8))
# sns.scatterplot(x = 'Time', y ='Raw', hue = 'Label', data = data)
```

```python
plt.figure(figsize=(18,6))
plt.plot(data['Time'], data['Raw'], label='ECG Signal')
plt.xlabel(r'Time (s)')
plt.ylabel('Raw (mV)')
plt.title("ECG Data")
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.plot(data['Time'], data['Raw'], label='ECG Signal')
plt.xlabel('Time (s)')
plt.ylabel('Raw (mV)')
plt.xlim(0,2)
plt.title("Close up of ECG Data")
plt.show()
```

```python
peak = data[data['Label']==1]
not_peak = data[data['Label']==0]
print("Peak : {}".format(len(peak)))
print("Not Peak : {}".format(len(not_peak)))
```

```python
LABELS = ['Not Peak', 'Peak']
count_labels = pd.value_counts(data['Label'], sort = True)
ax = count_labels.plot(kind = 'bar', rot = 0)
for idx, count in enumerate(count_labels):
    ax.text(idx, count + 100, str(count), ha='center', va='bottom')

plt.title('Label Distribution')
plt.xticks(range(2), LABELS)
plt.xlabel('Label')
plt.ylabel('Frequency')
```
